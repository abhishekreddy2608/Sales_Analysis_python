# Graph

# Gender

```python
dt.columns
```

```python
sns.countplot(x='Gender',data=dt)
```

```python
ax=sns.countplot(x='Gender',data=dt)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=dt.groupby('Gender')['Amount'].sum().sort_values(ascending=False)
sales_gen
```

```python
# Creating a bar plot
bx=sns.barplot(x=sales_gen.index, y=sales_gen.values)
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.title('Total Amount by Gender')
plt.show()
```

#In Above Graph female have more purchasing power as compared to men

# State

```python
dt.columns
```

```python
State_orders=dt.groupby('State')['Orders'].sum().sort_values(ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(30,5)})
```

```python
sns.barplot(x=State_orders.index, y=State_orders.values)
plt.xlabel('State')
plt.ylabel('Orders')
plt.title('Total Orders by State ')
plt.show()
```

From the Above Graph most of orders came from uttar pradesh and Maharashtra

```python
sales_amount=dt.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize':(30,5)})
```

```python
sns.barplot(x=sales_amount.index, y=sales_amount.values)
plt.xlabel('State')
plt.ylabel('Orders')
plt.title('Total Orders by State')
plt.show()
```

```python
purchase=dt.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(5)
```

From the Above graphs we can see that most of the sold products are from Food,Clothing & Apparel, Electronic & Gadgets.
